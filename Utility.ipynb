```python
import numpy as np
n_samples = 2000
amin, amax = -0.2, 0.2
samples = np.zeros((0,))    # empty for now
while samples.shape[0] < n_samples:
    s = np.random.normal(0, 0.2/3, size=(n_samples,))
    accepted = s[(s >= amin) & (s <= amax)]
    samples = np.concatenate((samples, accepted), axis=0)
samples = samples[:n_samples]
```

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 13})
mu=0
sigma=1/3
count, bins, ignored = plt.hist(samples, bins=100)
#plt.plot(bins, 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(- (bins - mu) ** 2 / (2 * sigma ** 2))*45, linewidth=2, color='r')
#t = np.linspace(-2, 3, 500)
#plt.plot(t, norm.pdf(t)/(norm.cdf(amax) - norm.cdf(amin)), 'r')
plt.grid()
plt.savefig('normal_trunc.pdf')
plt.show()
```
